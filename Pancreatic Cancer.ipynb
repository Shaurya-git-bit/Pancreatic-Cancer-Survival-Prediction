#              MACHINE LEARNING

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
from sklearn.metrics import confusion_matrix, f1_score, precision_score, recall_score,accuracy_score
```

```python
df_encoded['Survival_Status'].value_counts(normalize=True)#highly imbalanced data
```

```python
X = df_encoded.drop(["Survival_Status", "Survival_Time_Months","Country"], axis=1) #dropping unnecessary columns and the columns that can cause leakage
y = df_encoded["Survival_Status"] #target (for prediction)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y, random_state=42) # splitting dataset into training and testing(80% training and 20 % testing)
```

```python
#Using SMOTE as dataset is highly imbalanced 
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)#resampling X_train and y_train to balance minority and mojority class using synthetic data
```

```python
#Using standard scaler 
scaler = StandardScaler()#creating object for standard scaler
X_train_lr = scaler.fit_transform(X_train_res)  # training and trasnforming features on data resampled using SMOTE 
X_test_lr = scaler.transform(X_test)           #transforming test data
```

# ========>LOGISTIC REGRESSION<========

```python
log_reg=LogisticRegression(max_iter=1000, random_state=42)#creating object for logistic regression

log_reg.fit(X_train_lr, y_train_res)   # Train model on SMOTE and scaled training data

y_pred_log_reg = log_reg.predict(X_test_lr)  #storing predicted values in pred_log_reg
```

```python
cmat_log_reg=confusion_matrix(y_test, y_pred_log_reg) # confusion matrix for logistic regression
f1_log_reg = f1_score(y_test,y_pred_log_reg)           # F1 score for logistic regression
prec_log_reg = precision_score(y_test, y_pred_log_reg) # precision logistic regression
rec_log_reg= recall_score(y_test, y_pred_log_reg)      #recall

print("Logistic Regression \n")
print("Accuracy:", accuracy_score(y_test,  y_pred_log_reg))
print("\n")
print("Confusion Matrix:\n", cmat_log_reg)
print("F1 Score:",f1_log_reg)
print("Precision: ",prec_log_reg)
print("Recall: ",rec_log_reg)
```

# ======>RANDOM FOREST<======

```python
rf = RandomForestClassifier(class_weight='balanced',random_state=42)
rf.fit(X_train_res, y_train_res)     #No training on scaled data as random forest is tree based
y_pred_rf = rf.predict(X_test)       #storing predictions in y_pred_rf
```

```python
cmat_rf=confusion_matrix(y_test, y_pred_rf)  ## confusion matrix for random forest
f1_rf=f1_score(y_test, y_pred_rf)             # F1 score for random forest
prec_rf=precision_score(y_test, y_pred_rf)     # precision for random forest
rec_rf=recall_score(y_test, y_pred_rf)         # recall for random forest

print("Random Forest Claissifier \n")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("\n")
print("Confusion Matrix:\n", cmat_rf)
print("F1 Score:",f1_rf)
print("Precision: ",prec_rf)
print("Recall: ",rec_rf)
```

# ======>XGBOOST<======

```python
xgb = XGBClassifier(random_state=42)                         #creating object for XGBOOST
xgb.fit(X_train_res, y_train_res)                            # training on balanced(SMOTE) data
y_pred_xgb = xgb.predict(X_test)                             # storing predicted data in y_pred_xgb
```

```python
cmat_xgb=confusion_matrix(y_test, y_pred_xgb) # confusion matrix for XGBOOST
f1_xgb=f1_score(y_test, y_pred_xgb)           # F1 score for XGBOOST
prec_xgb=precision_score(y_test, y_pred_xgb)  # precision for XGBOOST
rec_xgb=recall_score(y_test, y_pred_xgb)      # recall for XGBOOST

print("XGBoost Claissifier \n")
print("Accuracy:", accuracy_score(y_test, y_pred_xgb))
print("\n")
print("Confusion Matrix:\n", cmat_xgb)
print("F1 Score:",f1_xgb)
print("Precision: ",prec_xgb)
print("Recall: ",rec_xgb)
```

# ----------HYPERPARAMETER TUNING----------

### Tuning for Random Forest

```python
rf_param_grid = {
    "n_estimators": [100, 200, 300, 400, 500],  #number of trees in model
    "max_depth": [None, 10, 20, 30, 40],        #Maximum depth of each tree.
    "min_samples_split": [2, 5, 10],  }         # Minimum samples needed to split a node
```

```python
from sklearn.model_selection import RandomizedSearchCV

rf_random = RandomizedSearchCV(
    estimator=rf,    # initial model used for tuning       
    param_distributions=rf_param_grid,  #dictionary of hyperparameters to try
    n_iter=20,             #number of iterations
    cv=3,                  # Number of folds for cross-validation
    verbose=1,             #live output shown during fitting
    random_state=42,       
    n_jobs=1 )              # Number of CPU cores to use
```

```python
rf_random.fit(X_train_res, y_train_res)
#fitting data from SMOT using best parameters available
```

```python
best_rf=rf_random.best_estimator_ #best parameters
best_pred_rf = best_rf.predict(X_test)  #best predicitons
```

```python
print("Best Parameters:", rf_random.best_params_)
```

```python
from sklearn.metrics import precision_recall_curve

# Get prediction probabilities
y_proba = best_rf.predict_proba(X_test)[:, 1]

# optimal threshold using F1-score
precision, recall, thresholds = precision_recall_curve(y_test, y_proba)
f1_scores = 2 * (precision * recall) / (precision + recall +1e-8)
optimal_threshold = thresholds[np.argmax(f1_scores)]

# Predict with optimal threshold
y_pred_optimal = (y_proba >= optimal_threshold).astype(int)

print("Random forest Optimal Threshold:", round(optimal_threshold, 3))
print("Precision:", precision_score(y_test, y_pred_optimal))
print("Recall:", recall_score(y_test, y_pred_optimal))
print("F1-Score:", f1_score(y_test, y_pred_optimal))
```
